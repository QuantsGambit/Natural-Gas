# Natural Gas Analysis

### Import Analysis
 - Variables Used
     - Import Prices
     - LNG Import Volume
     
### Hypothesis 
### - The variables will be highly correllated with lng import volume being the independent variable, and import prices as the dependent variable. As our lng import volume increases our import prices then increase, and vice versa.

```python
#----------------------------------------------------------------------------------
# Importing Modules
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
from matplotlib import dates as mpl_dates
#----------------------------------------------------------------------------------
```

```python
#----------------------------------------------------------------------------------
# Importing data
data = pd.read_csv('data0.csv')
#----------------------------------------------------------------------------------
```

```python
#----------------------------------------------------------------------------------
# Cleaning data
x0 = data['Dates']
y0 = data['Import Prices']
y1 = data['Import Volume As LNG']

x0_ = x0[-13:]
y0_ = y0[-13:].astype(float)
y1_ = y1[-13:].astype(float)

y1_Mean = sum(y1_) / len(y1_)
y1_Mean_x = ['2019-04-1', '2020-04-1']
y1_Mean_y = [y1_Mean, y1_Mean]
#----------------------------------------------------------------------------------
```

```python
#----------------------------------------------------------------------------------
# Plotting Import prices vs LNG Import Volume
fig, ax1 = plt.subplots()
c1 = 'teal'
ax1.set_ylabel('Import Prices', color=c1)
x0e = ax1.plot(x0_, y0_, color=c1, label='Import Prices')
ax1.tick_params(axis='y', labelcolor=c1)
ax1.format_xdata = mdates.DateFormatter('%Y-%m-%d')
#----------------------------------------------------------------------------------
ax2 = ax1.twinx()
c2 = 'firebrick'
ax2.set_ylabel('LNG Import Volume', color=c2)
x0f =ax2.plot(x0_, y1_, color=c2, label='LNG Import Volume')
ax2.tick_params(axis='y', labelcolor=c2)
ax2.format_xdata = mdates.DateFormatter('%Y-%m-%d')
#----------------------------------------------------------------------------------
ax3 = ax2.twiny()
c3 = 'darkmagenta'
x0i = ax3.plot(y1_Mean_x, y1_Mean_y, color=c3, label='LNG Import Volume Mean', linestyle='--')
ax3.get_xaxis().set_visible(False)
#----------------------------------------------------------------------------------
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title("Import Prices vs Import Volume")
plt.grid(True)
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%Y-%m-%d')
plt.gca().xaxis.set_major_formatter(date_format)
test = x0e+x0f+x0i
test1 = [t.get_label() for t in test]
ax1.legend(test, test1, loc=0)
plt.show()
#----------------------------------------------------------------------------------
```

### Notes
 - Automatically see a correlation as expected, however i thought the lng import volume would be the lead indicator as import prices lagged behind it.
 
 
 - We see lng import volume return to the mean before catching up to import prices, but when it cought up import prices started falling sending lng import volume back to it's mean.
 
 
 - Import prices vs lng import volume is growing a big spread, last time this happened volume reached back to mean before catching up to prices.
