# Natural Gas Forawrds Curve
A Forwards curve is the expected prices of asset $X$ at future times $T_n$

Natural Gas is a seasonal commodity, where prices hit highs in the winter (cold weather) and bottom in the summer (hot weather)

Traders price this in years ahead, via futures contracts which is what we use to derive the forwards curve.

### Lets start with importing our dataset:

```python
#--------------------------------------
# Import Data
data = pd.read_csv('./Data/forwards_curve.csv') 
data.head()
#--------------------------------------
```

### Now lets plot our Forwards Curve:

```python
#--------------------------------------
# Plotting
x = data['Date']
y = data['Price']
figure(figsize=(15,6),dpi=80)
plt.plot(x,y)
plt.ylabel("Price")
ax = plt.gca()
ax.tick_params(axis='x',rotation = 45)
#--------------------------------------
```

### Result:
- We can clearly see how traders price in the season's to Natural Gas Markets
    - Higher prices going into the winter
    - Lower prices going into the summer
